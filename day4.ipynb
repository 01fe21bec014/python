```python
def add(*args):
  sum=0
  for i in  args:
    sum+= i
  return sum
print(add(1,2,3,4,5))
```

```python
import random

gamelist=['rock','paper','scissors']
comp=random.choice(gamelist)
user=input("enter either rock or paper or scissors")
print(comp)
print(user)


if(comp=="rock" and user =="paper"):
  print("computer won")
elif(comp=="paper" and user =="scissors"):
  print("you won")
elif(comp=="scissors" and user =="paper"):
  print("computer won")
else:
  print("TIE")
```
